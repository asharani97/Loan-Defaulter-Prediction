```python
cat_features = [f for f in buerau_merge_bal.columns if buerau_merge_bal[f].dtype == 'object']
categorical_var = [x for x in buerau_merge_bal.columns if x in cat_features + ['SK_ID_CURR']]
cat_features
```

```python
continuous_var = [x for x in buerau_merge_bal.columns if x not in cat_features]
continuous_var
```

```python
categorical = pd.get_dummies(buerau_merge_bal[cat_features])
categorical['SK_ID_CURR'] = buerau_merge_bal['SK_ID_CURR']

categorical.head(5)
```

```python
cat_group_df = categorical.groupby('SK_ID_CURR').agg('sum').reset_index()
cat_group_df.head(10)
```

```python
cat_group_df.shape
```

```python
def agg_numeric(df, group_var):

    # Remove id variables other than grouping variable
    for col in df:
        if col != group_var and 'SK_ID' in col:
            df = df.drop(columns = col)
            
    group_ids = df[group_var]
    df[group_var] = group_ids

    # Group by the specified variable and calculate the statistics
    agg = df.groupby(group_var).agg('mean').reset_index()

    return agg
```

```python
df=buerau_merge_bal[continuous_var]
```

```python
df.shape
```

```python
num_group_df = agg_numeric(df.drop(columns = ['SK_ID_BUREAU']), group_var = 'SK_ID_CURR')
num_group_df.head()
```

```python
buerau_balance_agg =pd.merge(num_group_df,cat_group_df,on='SK_ID_CURR')
```
