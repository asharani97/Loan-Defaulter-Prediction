```python
# numpy and pandas for data manipulation
import numpy as np
import pandas as pd 

# sklearn preprocessing for dealing with categorical variables
from sklearn.preprocessing import LabelEncoder

# File system manangement
import os
from scipy import stats

# Suppress warnings 
import warnings
warnings.filterwarnings('ignore')
```

```python
# Testing data features
app_test = pd.read_csv('F:/Project/application_test.csv')
print('Testing data shape: ', app_test.shape)
app_test.head()
```

```python
# Bureau data features
bureau = pd.read_csv('F:/Project/bureau.csv')
print('bureau data shape: ', bureau.shape)
bureau.head()
```

```python
# Bureau data features
bureau_balance = pd.read_csv('F:/Project/bureau_balance.csv')
print('bureau_balance data shape: ', bureau_balance.shape)
bureau_balance.head()
```

```python
# joining bureau and bureau_balance
bureau = bureau.merge(bureau_balance, how='left', left_on='SK_ID_BUREAU', right_index=True)
```

```python
bureau
```

```python
bureau.drop(['SK_ID_BUREAU_x'], axis=1, inplace=True)
bureau.drop(['SK_ID_BUREAU_y'], axis=1, inplace=True)
```

```python
# Splitting categorical and numeric in bureau
cat_features = [f for f in bureau.columns if bureau[f].dtype == 'object']
```

```python
categorical_var = [x for x in bureau.columns if x in cat_features + ['SK_ID_CURR']]
```

```python
continuous_var = [x for x in bureau.columns if x not in categorical_var]
```

```python
bureau_1 = bureau.groupby('SK_ID_CURR').agg({k: lambda x: stats.mode(x)[0] for k in categorical_var})
```

```python
bureau_1.head()
```

```python
bureau_2 = bureau.groupby('SK_ID_CURR').agg({k:lambda x: x.mean() for k in continuous_var})
```

```python
bureau_2.head()
```

```python
bureau=pd.merge(bureau_1, bureau_2, left_index=True, right_index=True)
```

```python
bureau.head()
```

```python
bureau.to_csv('mearged_bureau_bureau_bal.csv', index = False)
```
