```python
import pandas as pd
import io
import numpy as np
import os
import matplotlib.pyplot as plt
import seaborn as sns
import pickle
from scipy import stats
print ('Ran')
```

```python
pos= pd.read_csv('C:\\home-credit-default-risk\\POS_CASH_balance.csv')
```

```python
cat_features = [f for f in pos.columns if pos[f].dtype == 'object']
```

```python
categorical_var = [x for x in pos.columns if x in cat_features + ['SK_ID_CURR']]
```

```python
continuous_var = [x for x in pos.columns if x not in categorical_var]
```

```python

agg_pos_cash_1 = pos.groupby('SK_ID_CURR').agg({k: lambda x: stats.mode(x)[0] for k in categorical_var})
```

```python
agg_pos_cash_1.head()
```

```python
agg_pos_cash_2 = pos.groupby('SK_ID_CURR').agg({k:lambda x: x.mean() for k in continuous_var})
```

```python
agg_pos_cash_2.head()
```

```python
agg_pos_cash_1.head(20)
```

```python
agg_pos_cash_2.head(20)
```

```python
pos=pd.merge(agg_pos_cash_1, agg_pos_cash_2, left_index=True, right_index=True)
```
