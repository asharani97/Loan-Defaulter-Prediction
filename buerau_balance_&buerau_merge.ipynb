```python
buerau_merge_bal=pd.merge(buearu,buearu_bal,on='SK_ID_BUREAU',how='left')
```

```python
cat_features = [f for f in buerau_merge_bal.columns if buerau_merge_bal[f].dtype == 'object']
categorical_var = [x for x in buerau_merge_bal.columns if x in cat_features + ['SK_ID_BUREAU']]
continuous_var = [x for x in buerau_merge_bal.columns if x not in categorical_var]
```

```python
# merging of categoical of buerau bal
agg_buerau_bal_1 = buerau_merge_bal.groupby(['SK_ID_CURR','SK_ID_BUREAU']).agg({k: lambda x: stats.mode(x)[0] for k in categorical_var})

# merging of numerical of buerau bal
agg_buerau_bal_2 = buerau_merge_bal.groupby(['SK_ID_CURR','SK_ID_BUREAU']).agg({k:lambda x: x.mean() for k in continuous_var})
agg_buerau_bal_2.shape
```

```python
agg_buerau_bal_1.head(20)
```

```python
agg_buerau_bal_2.head(20)
```

```python
# merging of categorical & numerical of buerau balance and buerau data
buerau_balance =pd.merge(agg_buerau_bal_1,agg_buerau_bal_2, left_index=True, right_index=True)
```

```python
buerau_balance.to_csv("C:\\home-credit-default-risk\\result\\buerau_bureau_bal_merge.csv",index=False)
```
