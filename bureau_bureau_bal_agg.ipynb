```python
# numpy and pandas for data manipulation
import numpy as np
import pandas as pd 

# sklearn preprocessing for dealing with categorical variables
from sklearn.preprocessing import LabelEncoder

# File system manangement
import os
from scipy import stats

# Suppress warnings 
import warnings
warnings.filterwarnings('ignore')
```

```python
# Bureau data features
bureau = pd.read_csv('F:/Project/Input_data/bureau.csv')
print('bureau data shape: ', bureau.shape)
bureau.head()
```

```python
# Bureau data features
bureau_balance = pd.read_csv('F:/Project/Input_data/bureau_balance.csv')
print('bureau_balance data shape: ', bureau_balance.shape)
bureau_balance.head()
```

```python
# joining bureau and bureau_balance
bureau_mearged = bureau.merge(bureau_balance, on='SK_ID_BUREAU', how='left')
bureau_mearged.head()
```

```python
bureau_mearged.to_csv('bureau_bureau_bal_m1.csv', index = False)
```

```python
bureau_mearged.columns
```

```python
# Groupby the client id (SK_ID_CURR), count the number of previous loans, and rename the column
previous_loan_counts = bureau_mearged.groupby('SK_ID_CURR', as_index=False)['SK_ID_BUREAU'].count().rename(columns = {'SK_ID_BUREAU': 'previous_loan_counts'})
previous_loan_counts.head()
```

```python
previous_loan_counts.shape
```

```python
categorical = pd.get_dummies(bureau_mearged.select_dtypes('object'))
categorical['SK_ID_CURR'] = bureau_mearged['SK_ID_CURR']
categorical.head()
```

```python
categorical_grouped = categorical.groupby('SK_ID_CURR').agg('sum')
categorical_grouped.head()
```

```python
continuous_var = [f for f in bureau_mearged.columns if bureau_mearged[f].dtype != 'object']
```

```python
continuous_grouped = bureau_mearged.groupby('SK_ID_CURR').agg({k:lambda x: x.mean() for k in continuous_var})
```

```python
print(categorical_grouped.shape)
print(continuous_grouped.shape)
```

```python
# merge aggregated numerical and categorical columns
bureau_new = categorical_grouped.merge(continuous_grouped, how='left', on='SK_ID_CURR')
```

```python
bureau_new
```

```python
# merge aggregated numerical and categorical columns
bureau_final = bureau_new.merge(previous_loan_counts, how='left', on='SK_ID_CURR', right_index=True)
```

```python
bureau_final.shape
```

```python
bureau_final.to_csv('bureau_bureau_bal_aggregate.csv', index = False)
```
